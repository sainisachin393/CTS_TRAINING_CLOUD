```python
#performing hot enconding
life_expectancy_df=pd.get_dummies(life_expectancy_df,columns=['Status'])
```

```python
# since most of the column is continous so filling with mean
life_expectancy_df=life_expectancy_df.apply(lambda x: x.fillna(x.mean()),axis=0)
```

```python
#creating test and train data
X=life_expectancy_df.drop(columns=['Life expectancy '],axis=1)
Y=life_expectancy_df[['Life expectancy ']]
```

```python
X=np.array(X).astype('float32')
Y=np.array(Y).astype('float32')
```

```python
# Splitting the data and testing using sklearn library
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
X_test,X_val,Y_test,Y_val=train_test_split(X_test,Y_test,test_size=0.5)
```

```python
train_data = pd.DataFrame({'Target': Y_train[:,0]})
for i in range(X_train.shape[1]):
    train_data[i] = X_train[:,i]
```

```python
val_data = pd.DataFrame({'Target': Y_val[:,0]})
for i in range(X_val.shape[1]):
    val_data[i] = X_val[:,i]
```

```python
val_data.head(5)
```

```python
# saving train and test data as csv file
train_data.to_csv('train.csv', header = False, index = False)

val_data.to_csv('validation.csv', header = False, index = False)

```
